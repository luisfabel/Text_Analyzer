# Análisis de Frases

```python
import pandas as pd
from io import StringIO
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
df=pd.DataFrame(tokenized_text,columns=['Frases'])
```

```python
vader = SentimentIntensityAnalyzer()
columns = ['Frases']
reporte = pd.DataFrame(df, columns=columns)
pd.set_option('max_colwidth',200)
scores = reporte['Frases'].apply(vader.polarity_scores).tolist()
scores_df = pd.DataFrame(scores)
reporte = reporte.join(scores_df, rsuffix='_right')
```

```python
reporte
```

```python
positivos=reporte.sort_values(by=['pos'],ascending=[False])
positivos.head(60)
```

```python
negativos=reporte.sort_values(by=['neg'],ascending=[False])
negativos.head(50)
```

```python
reporte.hist('compound',figsize=(10,5))
```

```python
reporte.compound.value_counts(bins=3)
```

```python
rpt=reporte.compound.value_counts(bins=3)
rpt.plot.bar(x ='Sentimiento', y='Calificación')
positions = (0, 1, 2)
labels = ("Positivo", "Neutral", "Negativo")
plt.xticks(positions, labels)
plt.show()
```

```python
Sentiment_count=reporte.groupby('compound').count()
plt.figure(figsize=(10,5))
plt.plot(Sentiment_count.index.values, Sentiment_count['Frases'])
plt.xlabel('<---Negativo -- Neutral -- Positivo--->')
plt.show()
```
